```r

#Устанавливаем библиотеки
install.packages("readxl")
install.packages("ggplot2")
install.packages("car")
install.packages("lmtest")

library(readxl)
library(ggplot2)
library(car)
library(lmtest)

#Подготовка
if (!dir.exists("plots")) dir.create("plots")
df <- read_excel("HR_Employees.xlsx", sheet = 1)

#1
cat("\n 1. Описательная статистика \n")
str(df)
summary(df[, c("Age", "MonthlyIncome", "TotalWorkingYears", "YearsAtCompany")])
table(df$Gender)
table(df$Education)
table(df$Attrition)

cat("Выводы: наблюдений", nrow(df), ", средний доход", round(mean(df$MonthlyIncome), 2),
    ", возраст от", min(df$Age), "до", max(df$Age), ", текучка",
    round(mean(df$Attrition == "Yes") * 100, 2), "%\n")

#2
cat("\n 2. Предикторы: Gender (кат), Education (кат), Age, TotalWorkingYears, YearsAtCompany (колич) \n")
df$Gender <- as.factor(df$Gender)
df$Education <- as.factor(df$Education)

#3
cat("\n 3. Корреляция \n")
cor(df[, c("MonthlyIncome", "Age", "TotalWorkingYears", "YearsAtCompany")], use = "complete.obs")

png("plots/scatter_quant.png", width = 800, height = 600)
par(mfrow = c(2, 2))
plot(df$Age, df$MonthlyIncome, main = "Age vs Income")
plot(df$TotalWorkingYears, df$MonthlyIncome, main = "TotalWorkYears vs Income")
plot(df$YearsAtCompany, df$MonthlyIncome, main = "YearsAtCompany vs Income")
dev.off()

png("plots/boxplot_gender.png")
boxplot(MonthlyIncome ~ Gender, df, main = "Income by Gender")
dev.off()
png("plots/boxplot_edu.png")
boxplot(MonthlyIncome ~ Education, df, main = "Income by Education")
dev.off()

#4
cat("\n 4. Корреляция между Age, TotalWorkingYears, YearsAtCompany \n")
print(cor(df[, c("Age", "TotalWorkingYears", "YearsAtCompany")]))
cat("Хи-квадрат Gender vs Education:\n")
print(chisq.test(table(df$Gender, df$Education)))

#5
cat("\n 5. Модель lm(MonthlyIncome ~ Age + TotalWorkingYears + YearsAtCompany + Gender + Education) \n")
model <- lm(MonthlyIncome ~ Age + TotalWorkingYears + YearsAtCompany + Gender + Education, data = df)
summary(model)

#6
cat("\n 6. Проверка допущений \n")
cat("VIF:\n")
print(vif(model))

res <- residuals(model)
png("plots/qqplot.png")
qqnorm(res)
qqline(res, col = "red")
dev.off()
cat("Шапиро-Уилка p =", shapiro.test(res)$p.value, "\n")

png("plots/resid_fitted.png")
plot(fitted(model), res, xlab = "Fitted", ylab = "Residuals")
abline(h = 0, col = "red")
dev.off()
cat("Бройша-Пагана p =", bptest(model)$p.value, "\n")

cooksd <- cooks.distance(model)
png("plots/cooks.png")
plot(cooksd, type = "h", main = "Cook's distance")
dev.off()
cat("Влиятельных наблюдений:", sum(cooksd > 4 / (nrow(df) - length(coef(model)) - 1)), "\n")

#7
cat("\n 7. Значимость и R² \n")
s <- summary(model)
cat("F =", s$fstatistic[1],
    "p =", pf(s$fstatistic[1], s$fstatistic[2], s$fstatistic[3], lower.tail = FALSE), "\n")
cat("R² =", round(s$r.squared, 4), "Adj R² =", round(s$adj.r.squared, 4), "\n")

#8
cat("\n 8. Интерпретация \n")
cat("Age:", round(coef(model)["Age"], 2), "(прирост за год)\n")
cat("TotalWorkingYears:", round(coef(model)["TotalWorkingYears"], 2), "(за год общего стажа)\n")
cat("YearsAtCompany:", round(coef(model)["YearsAtCompany"], 2), "(за год в компании)\n")
cat("GenderMale:", round(coef(model)["GenderMale"], 2), "(мужчины vs женщины)\n")
for (lev in levels(df$Education)[-1]) {
  cat("Education", lev, ":", round(coef(model)[paste0("Education", lev)], 2),
      "(относительно Below College)\n")
}

#*9
cat("\n 9. Стандартизованные коэффициенты \n")
# Ручная стандартизация
df_scaled <- df
df_scaled$Age <- scale(df$Age)
df_scaled$TotalWorkingYears <- scale(df$TotalWorkingYears)
df_scaled$YearsAtCompany <- scale(df$YearsAtCompany)
df_scaled$MonthlyIncome <- scale(df$MonthlyIncome)

model_scaled <- lm(MonthlyIncome ~ Age + TotalWorkingYears + YearsAtCompany + Gender + Education,
                   data = df_scaled)
beta <- coef(model_scaled)[-1]  # без интерсепта
print(beta)
cat("Наибольший по модулю:", names(which.max(abs(beta))), "\n")

#10
cat("\n 10. График предсказанных vs фактических \n")
pred <- fitted(model)
ggplot(data.frame(actual = df$MonthlyIncome, pred = pred), aes(pred, actual)) +
  geom_point(alpha = 0.3) +
  geom_abline(slope = 1, col = "red", linetype = 2) +
  labs(title = "Predicted vs Actual", x = "Predicted", y = "Actual") +
  theme_minimal()
ggsave("plots/pred_vs_actual.png", width = 6, height = 4)
```
